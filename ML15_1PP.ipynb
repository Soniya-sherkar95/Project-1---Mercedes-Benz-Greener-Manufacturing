1. If for any column(s), the variance is equal to zero, then you need to remove those variable(s)

3. Perform dimensionality reduction for 






```python
train_data_feature = train_df.drop(columns={'y','ID'})
train_data_target = train_df.y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95) # pca with 95%
pca.fit(train_data_feature,train_data_target)
train_data_feature_tran = pca.fit_transform(train_data_feature)
train_data_feature_tran.shape
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt
```

```python
train_x,test_x,train_y,test_y = train_test_split(train_data_feature_tran,train_data_target,test_size=.3,random_state=7)
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

Predict your test_df values using XGBoost.

```python
xgb_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.4, max_depth = 10, alpha = 6, 
                           n_estimators = 20)
model = xgb_reg.fit(train_x,train_y)
print('RMSE = ',sqrt(mean_squared_error(model.predict(test_x),test_y)))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pred_test_y = model.predict(test_x)

plt.figure(figsize=(10,5))

sns.distplot(test_y[test_y<160], color="skyblue", label="Actual value")
sns.distplot(pred_test_y[pred_test_y<160] , color="red", label="Predicted value")
plt.legend()

plt.tight_layout()
```
